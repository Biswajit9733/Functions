1. What is the difference between a function and a method in Python?

In Python, functions and methods are both blocks of reusable code, but they have some key differences.

A function is a block of code that is defined using the def keyword and can be called independently. Functions can exist outside of classes and are not tied to any object.

A method is a function that is associated with an object or class. It is defined inside a class and operates on the data contained in the object. Methods are called on objects and have access to the object’s internal state through the self parameter.

```python
# Function example
def greet(name):
    return f"Hello, {name}!"

# Calling the function
print(greet("Alice"))

# Method example
class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

# Creating an object of the Greeter class
greeter = Greeter()

# Calling the method on the object
print(greeter.greet("Bob"))
```

2.  Explain the concept of function arguments and parameters in Python.

In Python, parameters and arguments are related but distinct concepts used when defining and calling functions.

Parameters are variables that are defined in the function signature and act as placeholders for the values that will be passed into the function when it's called.

Arguments are the actual values that you pass to the function when calling it, which correspond to the parameters.

Python supports several types of arguments:

Positional arguments: Passed in order.
Keyword arguments: Passed with parameter names.
Default arguments: Parameters with default values.
Variable-length arguments: Allows for arbitrary numbers of arguments (*args, **kwargs).


```python
# Function with different types of parameters
def greet(greeting, name="User", *args, **kwargs):
    print(f"{greeting}, {name}!")

    # args example
    for extra in args:
        print(f"Additional argument: {extra}")

    # kwargs example
    for key, value in kwargs.items():
        print(f"{key}: {value}")

# Positional and default argument
greet("Hello")

# Positional and keyword arguments
greet("Hi", name="Alice")

# Variable-length arguments (*args) and keyword arguments (**kwargs)
greet("Greetings", "Bob", "Extra1", "Extra2", age=25, city="NY")
```

3. What are the different ways to define and call a function in Python?

In Python, there are several ways to define and call a function. Here are some common ones:

Normal function definition: The most common way to define a function using the def keyword.
Lambda functions: A short, anonymous function defined with the lambda keyword for small, single-use functions.
Default arguments: Functions can have default values for parameters.
Variable-length arguments: Using *args for positional arguments and **kwargs for keyword arguments to accept a variable number of arguments.
Nested functions: A function defined inside another function.
Calling functions with positional arguments: Arguments are passed in the same order as defined.
Calling functions with keyword arguments: Arguments are passed using parameter names.

```python
# 1. Normal function definition
def add(a, b):
    return a + b

# 2. Lambda function
multiply = lambda x, y: x * y

# 3. Function with default argument
def greet(name="Guest"):
    return f"Hello, {name}!"

# 4. Function with *args and **kwargs
def display_info(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)

# 5. Nested function
def outer_function(x):
    def inner_function(y):
        return x + y
    return inner_function(5)

# Calling the functions
print(add(2, 3))  # Positional arguments
print(multiply(2, 3))  # Lambda function
print(greet())  # Default argument
print(greet("Alice"))  # Keyword argument
display_info(1, 2, 3, name="Alice", age=30)  # *args and **kwargs
print(outer_function(10))  # Nested function call
```

4. What is the purpose of the `return` statement in a Python function?

The return statement in a Python function is used to exit the function and send a value back to the caller. It allows the function to produce an output, which can then be stored in a variable or used in further operations. Without a return statement, the function will return None by default.

Key points:

The return statement can return any data type (e.g., integers, strings, lists, dictionaries, etc.).
A function can return multiple values as a tuple by separating them with commas.
The function terminates immediately when a return statement is executed, even if there are more lines of code afterward.
If no return statement is used, the function implicitly returns None.
You can use return without any value to exit the function early.

```python
# Function using return
def add(a, b):
    return a + b  # Returns the sum of a and b

# Function returning multiple values
def get_coordinates():
    x, y = 10, 20
    return x, y  # Returns a tuple (x, y)

# Function with early return
def check_even(num):
    if num % 2 == 0:
        return "Even"
    return "Odd"

# Calling the functions
result = add(5, 3)
print(result)

x, y = get_coordinates()
print(x, y)

print(check_even(4))
print(check_even(7))
```

5.  What are iterators in Python and how do they differ from iterables?

In Python, iterables and iterators are related but have key differences:

Iterable: An object that can be looped over (iterated through). Examples include lists, tuples, strings, and dictionaries. An iterable must implement the __iter__() method, which returns an iterator.

Iterator: An object that represents a stream of data. It is an object that can be iterated upon and produces values one at a time when requested. An iterator must implement two methods: __iter__() (to return itself) and __next__() (to fetch the next value). Once all items are exhausted, StopIteration is raised.

Differences:
An iterable can be converted to an iterator using the iter() function.
An iterator fetches values using the next() function, whereas an iterable can only be looped using a for loop.

```python
# Iterable example: A list is iterable
my_list = [1, 2, 3]

# Getting an iterator from the iterable
my_iterator = iter(my_list)

# Iterator: Fetching items one by one using next()
print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))

# If we try to go further, StopIteration is raised
# print(next(my_iterator))  # Uncommenting this line will raise StopIteration

# Using an iterable in a for loop
for item in my_list:
    print(item)
```

6.  Explain the concept of generators in Python and how they are defined.

In Python, generators are a type of iterator that allow you to iterate over data one item at a time, without having to store the entire sequence in memory at once. This makes them memory-efficient for handling large datasets or infinite sequences. Generators are defined using the yield keyword instead of return.

Key points:

A generator is defined like a normal function but uses yield to return values one at a time.
When a generator function is called, it doesn’t run immediately but returns a generator object.
Each time the generator’s __next__() method (or next() function) is called, the function runs until it hits a yield statement, pauses, and returns the yielded value.
The state of the function (local variables, execution point) is saved, so the function resumes from where it left off on the next call.
Generators are useful for working with large data or infinite streams of data where you don’t want to load everything into memory at once.


```python
# Generator function using 'yield'
def countdown(n):
    while n > 0:
        yield n  # Yield the current value of n
        n -= 1

# Creating a generator object
gen = countdown(5)

# Iterating through the generator
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))

# StopIteration will be raised after the last item is exhausted
# print(next(gen))  # Uncommenting this line will raise StopIteration

# Using a generator in a loop
for number in countdown(3):
    print(number)
```

7. What are the advantages of using generators over regular functions?

Generators offer several advantages over regular functions, especially when dealing with large data or streams of values:

Memory Efficiency: Generators produce values one at a time using the yield statement, rather than generating all values at once and storing them in memory. This makes them ideal for processing large datasets or infinite sequences without consuming large amounts of memory.

Lazy Evaluation: Values are produced only when needed, rather than all at once. This is known as lazy evaluation, which can improve performance by avoiding unnecessary computation and memory use.

State Persistence: A generator automatically saves its state (local variables and the current point of execution) between each call to next(), allowing it to resume from where it left off. This is useful for complex iteration logic.

Simpler Code: Generators allow you to write more concise and readable code for generating sequences compared to using complex logic with loops and storing intermediate results.

Infinite Sequences: Generators can handle infinite sequences, like a stream of real-time data, without exhausting memory, whereas regular functions would require storing all values in a list.

```python
# Regular function generating a list of squares
def square_list(n):
    return [i**2 for i in range(n)]

# Generator function yielding squares lazily
def square_generator(n):
    for i in range(n):
        yield i**2

# Regular function consumes memory for large n
squares = square_list(1000000)

# Generator only yields values as needed, saving memory
gen_squares = square_generator(1000000)

# Accessing generator values one by one
print(next(gen_squares))
print(next(gen_squares))
```

8. What is a lambda function in Python and when is it typically used?

A lambda function in Python is an anonymous, small, and inline function defined using the lambda keyword. It can take any number of arguments but can only have one expression, which is evaluated and returned. Lambda functions are typically used when a simple function is required for a short period and defining a full function using def would be overkill.

Key points:

Syntax: lambda arguments: expression
They are often used as arguments to higher-order functions like map(), filter(), and sorted().
Lambda functions are useful when a small, one-off function is required for a specific task.
Since they are restricted to a single expression, lambda functions are concise but limited in complexity compared to regular functions.

```python
# Regular function
def add(a, b):
    return a + b

# Equivalent lambda function
add_lambda = lambda a, b: a + b

# Using the lambda function
print(add_lambda(2, 3))

# Example with map() using lambda to square numbers
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x ** 2, numbers))
print(squared)

# Sorting a list of tuples by the second element using lambda
tuples = [(1, 2), (3, 1), (5, 4)]
sorted_tuples = sorted(tuples, key=lambda x: x[1])
print(sorted_tuples)
```

9. Explain the purpose and usage of the `map()` function in Python.

The map() function in Python is a built-in function used to apply a specified function to each item in an iterable (like a list, tuple, or string) and returns an iterator (a map object). This allows for efficient and concise transformations of data without the need for explicit loops.
Key points:

The function argument can be a regular function, a lambda function, or any callable object.
map() can also take multiple iterables as arguments, applying the function in parallel.
The output from map() is an iterator, so it needs to be converted to a list or another iterable to view the results.
Using map() can make code cleaner and more readable by reducing the need for loops.


```python
# Function to double a number
def double(x):
    return x * 2

# Using map() with a function
numbers = [1, 2, 3, 4, 5]
doubled_numbers = list(map(double, numbers))
print(doubled_numbers)

# Using map() with a lambda function
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)

# Using map() with multiple iterables
list1 = [1, 2, 3]
list2 = [4, 5, 6]
summed_lists = list(map(lambda x, y: x + y, list1, list2))
print(summed_lists)
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

The map(), reduce(), and filter() functions in Python are all higher-order functions used for processing iterables, but they serve different purposes:

map(): This function applies a specified function to each item in an iterable (like a list) and returns an iterator with the results. It is useful for transforming data. For example, you can use map() to square each number in a list.

filter(): This function applies a predicate (a function that returns True or False) to each item in an iterable and returns an iterator with items for which the predicate is True. It is used for selecting items based on a condition. For example, you can use filter() to get only the even numbers from a list.

reduce(): This function is part of the functools module and applies a binary function cumulatively to the items in an iterable, reducing them to a single value. It is useful for aggregating data, such as calculating the product of all numbers in a list.

```python
from functools import reduce

# Sample data
numbers = [1, 2, 3, 4, 5]

# Using map() to square each number
squared = list(map(lambda x: x ** 2, numbers))
print("Squared:", squared)

# Using filter() to get even numbers
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print("Even Numbers:", even_numbers)

# Using reduce() to calculate the product of all numbers
product = reduce(lambda x, y: x * y, numbers)
print("Product:", product)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
# Function to check if a number is prime
def is_prime(n):
    if n <= 1:  # 0 and 1 are not prime numbers
        return False
    for i in range(2, int(n**0.5) + 1):  # Check for factors up to the square root of n
        if n % i == 0:
            return False
    return True

# Example usage
numbers_to_check = list(range(1, 201))  # List of numbers from 1 to 200
prime_numbers = [num for num in numbers_to_check if is_prime(num)]

# Displaying the result
print("Prime numbers from 1 to 200:", prime_numbers)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms  # Number of terms to generate
        self.current = 0    # Current term index
        self.a, self.b = 0, 1  # Starting values for Fibonacci sequence

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.terms:
            # Get the next Fibonacci number
            fib_number = self.a
            self.a, self.b = self.b, self.a + self.b  # Update values for next Fibonacci number
            self.current += 1
            return fib_number
        else:
            raise StopIteration  # Stop iteration when the limit is reached

# Example usage
num_terms = 10  # Specify the number of terms
fibonacci_sequence = FibonacciIterator(num_terms)

# Displaying the Fibonacci sequence
print("Fibonacci sequence:")
for number in fibonacci_sequence:
    print(number)
```

6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
# Generator function to yield powers of 2
def powers_of_two(exponent):
    for i in range(exponent + 1):  # Iterate from 0 to the given exponent
        yield 2 ** i  # Yield 2 raised to the power of i

# Example usage
exponent = 10  # Specify the maximum exponent
power_generator = powers_of_two(exponent)

# Displaying the powers of 2
print("Powers of 2 up to 2^{}:".format(exponent))
for power in power_generator:
    print(power)
```

7.  Implement a generator function that reads a file line by line and yields each line as a string.

```python
# Generator function to simulate reading a file line by line
def read_file_line_by_line(fake_file):
    for line in fake_file:
        yield line  # Yield each line

# Simulated file content as a list of strings (each string represents a line)
fake_file_content = [
    "This is the first line.",
    "Here is the second line.",
    "This is the third line.",
    "Finally, this is the fourth line."
]

# Example usage
line_generator = read_file_line_by_line(fake_file_content)

# Displaying each line
print("Reading lines from the simulated file:")
for line in line_generator:
    print(line)
```

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 37, 100]

# Using map() to convert the list of Celsius temperatures to Fahrenheit
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Displaying the result
print("Celsius temperatures:", celsius_temperatures)
print("Fahrenheit temperatures:", fahrenheit_temperatures)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
# Function to check if a character is not a vowel
def is_not_vowel(char):
    vowels = "aeiouAEIOU"
    return char not in vowels

# Input string
input_string = "Hello, World!"

# Using filter() to remove vowels
filtered_string = ''.join(filter(is_not_vowel, input_string))

# Displaying the result
print("Original string:", input_string)
print("String without vowels:", filtered_string)
```

11. Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

Write a Python program using lambda and map.

```python
# Sample data: a list of orders with (Order Number, Book Title, Quantity, Price per Item)
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

# Lambda function to calculate the total for each order
# Adds 10€ if the total is less than 100€
calculate_total = lambda order: (order[0], order[2] * order[3] if order[2] * order[3] >= 100 else order[2] * order[3] + 10)

# Using map() to apply the lambda function to each order
order_totals = list(map(calculate_total, orders))

# Displaying the result
print("Order totals with conditions applied:", order_totals)
```

Explanation:
The orders list contains sublists, where each sublist consists of:
Order number
Book title and author
Quantity of books ordered
Price per item
The calculate_total lambda function multiplies the quantity and price per item. If the result is less than 100€, it adds 10€.
map() applies this function to every sublist in orders.
The result is a list of tuples containing the order number and the calculated total.
